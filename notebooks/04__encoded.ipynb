```python
import pandas as pd
import numpy as np
#import sklearn.preprocessing.LabelBinarizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import OrdinalEncoder
import itertools
```

```python
X_COL =['YearsCodePro', 'OrgSize', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith',
       'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith',
       'PlatformHaveWorkedWith', 'PlatformWantToWorkWith',
       'WebframeHaveWorkedWith', 'WebframeWantToWorkWith',
       'MiscTechHaveWorkedWith', 'MiscTechWantToWorkWith',
       'ToolsTechHaveWorkedWith', 'ToolsTechWantToWorkWith',
       'NEWCollabToolsHaveWorkedWith', 'NEWCollabToolsWantToWorkWith']
```

```python
df = pd.read_pickle('../data/processed/02__cleaning.pkl')
df.reset_index(drop = 'index', inplace = True)

encoded_df = pd.read_csv('../data/processed/03__encoded.csv',index_col = 0)
# encoded_df.reset_index(drop = 'index', inplace = True)
# df
```

```python
df2 = df.copy()
encoded_df2 = encoded_df.copy()
```

```python
dic = {}
wantlist = []
for i in range(2, len(X_COL),2):
    have =  set(itertools.chain.from_iterable(df2[X_COL[i]].dropna().ravel().tolist()))
    want = set(itertools.chain.from_iterable(df2[X_COL[i + 1]].dropna().ravel().tolist()))
    
    
    wantlist.append(X_COL[i + 1])
    dic[X_COL[i]] = list(have.union(want))

```

```python
for key_values, want in zip(dic.items(),wantlist):
    encoded_df2[[f"{(key_values[0], key_values[1][i])}" for i in range(len(key_values[1]))]] =\
    encoded_df2[[f"{(key_values[0], key_values[1][i])}" for i in range(len(key_values[1]))]].values +\
    encoded_df2[[f"{(want, key_values[1][i])}" for i in range(len(key_values[1]))]].values * .5 
    
    encoded_df2.drop([f"{(want, key_values[1][i])}" for i in range(len(key_values[1]))],axis = 1, inplace = True)
```

```python
size_years = (encoded_df2['OrgSize'].values) + (encoded_df2['YearsCodePro'].values) + 1
```

```python
encoded_df3 = pd.DataFrame(size_years[:,np.newaxis] * encoded_df2.iloc[:,:-25].values,columns = encoded_df2.columns[:-25] )
encoded_df3.drop(['OrgSize','YearsCodePro'], axis =1, inplace = True)
```

```python
encoded_df2 = pd.merge(encoded_df3,encoded_df2.iloc[:,-25:],left_index=True, right_index=True)
columns = [eval(encoded_df2.columns[i]) for i in range(len(encoded_df2.columns))]
```

```python
encoded_df2.columns = pd.MultiIndex.from_tuples(columns)
```
